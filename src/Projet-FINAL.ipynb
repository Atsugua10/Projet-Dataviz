## Nettoyage

```python
# supression des collonnes a donees manquanes a plus de 90 %
df_sampled_2020= df_sampled_2020[df_sampled_2020.columns[df_sampled_2020.isna().sum()/df_sampled_2020.shape[0]<0.75]]
```

```python
df_sampled_2020.columns
```

```python
df_sampled_2020.info()
```

```python
# nettoyage des donnees manquantes
# pour ne pas pedre des odnnes j'affecte les valkeurs moyennes aux donnees manquantes dans 
#valeur foncieres, surface terrain et nomvre de pieces manquantes

df_sampled_2020['valeur_fonciere'] = df_sampled_2020['valeur_fonciere'].fillna(value=df['valeur_fonciere'].mean())
df_sampled_2020['surface_terrain'] = df_sampled_2020['surface_terrain'].fillna(value=df['surface_terrain'].mean())
df_sampled_2020['nombre_pieces_principales'] = df_sampled_2020['nombre_pieces_principales'].fillna(value=df['nombre_pieces_principales'].mean())



# suppression des lignes vides
for i in df_sampled_2020.columns:
    if (df_sampled_2020.isnull().values.any() == True):
        df_sampled_2020.dropna(axis = 0, inplace=True) 
```
